# Visualizing The Gender Gap in College Degrees

Disclaimer: This project was inspired and partly instructed by the guided project of dataquest.io.

The aim of this project is to visualize and analyze the gender gap in different college majors. The data in use comes from the Department of Education Statistics (USA) and contains the percentage of bachelor's degrees granted to women from 1970 to 2012. Randal Olson, a datascientist at the UNiversity of Pennsylvania, posted a cleaned version of this data to download from his personal website.

We're going to start of with some simple linecharts visualizing the different STEM-college majors. 


```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
# defining rgbvalues
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
# defining STEM-degrees
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

# creating the figure
fig = plt.figure(figsize=(18, 3))

# iterating over the indexes in the figure and creating subplots each iteration for each category
for sp in range(0,6):
    # defining subplot
    ax = fig.add_subplot(1,6,sp+1)
    # defining data input, two lines per plot
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    # hide axes spines
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    # set limitations
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # set title
    ax.set_title(stem_cats[sp])
    # set ticks
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    # set lables
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

## Comparing Across Different Categories

Next we're going to visualize the other college majors and group them into STEM, liberal arts and others.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16,20))

# iterateion for STEM
for index in range(0,18,3):
    # keep track of the list index
    cat_index = int(index/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set title
    ax.set_title(stem_cats[cat_index])
    # set text annotations
    if index == 0:
        ax.text(2002, 70, "Women")
        ax.text(2002, 28, "Men")
    if index == 15:
        ax.text(2002, 68, "Women")
        ax.text(2002, 28, "Men")

# iterateion for lib_arts
for index in range(1,16,3):
    # keep track of the list index
    cat_index = int((index-1)/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set title
    ax.set_title(lib_arts_cats[cat_index])
    # set text annotations
    if index == 1:
        ax.text(2002, 62, "Women")
        ax.text(2002, 35, "Men")
    
# iterateion for other
for index in range(2,20,3):
    # keep track of the list index
    cat_index = int((index-2)/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[other_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set title
    ax.set_title(other_cats[cat_index])
    # set text annotations
    if index == 2:
        ax.text(2002, 75, "Women")
        ax.text(2002, 23, "Men")
    if index == 17:
        ax.text(2002, 68, "Men")
        ax.text(2002, 28, "Women")
```

## Improving Readability

After removing the y-axis labels we're sort of lacking an point of reference in the plots. We're going to use a horizontal line at 50% to improve the readability of the chart.

```python
fig = plt.figure(figsize=(16,20))

# iterateion for STEM
for index in range(0,18,3):
    # keep track of the list index
    cat_index = int(index/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # set title
    ax.set_title(stem_cats[cat_index])
    # set text annotations
    if index == 0:
        ax.text(2002, 70, "Women")
        ax.text(2002, 28, "Men")
    if index == 15:
        ax.text(2002, 68, "Women")
        ax.text(2002, 28, "Men")
    # remove all spines
    # hiding spines using the axis.spines.Spine class
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    # removing the axis labels except the bottom-most one
    if index < 15:
        ax.tick_params(bottom=False, top=False, right=False, left=False, labelbottom=False)
    else:
        ax.tick_params(bottom=False, top=False, right=False, left=False)
        # reset bottom spine for the last row
        ax.spines["bottom"].set_visible(True)
    # add a horizontal line
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)



# iterateion for lib_arts
for index in range(1,16,3):
    # keep track of the list index
    cat_index = int((index-1)/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # set title
    ax.set_title(lib_arts_cats[cat_index])
    # set text annotations
    if index == 1:
        ax.text(2002, 62, "Women")
        ax.text(2002, 35, "Men")
    # remove all spines
    # hiding spines using the axis.spines.Spine class
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    # removing the axis labels except the bottom-most one
    if index < 13:
        ax.tick_params(bottom=False, top=False, right=False, left=False, labelbottom=False)
    else:
        ax.tick_params(bottom=False, top=False, right=False, left=False)
        # reset bottom spine for the last row
        ax.spines["bottom"].set_visible(True)
    # remove y-labels
    ax.tick_params(labelleft=False)
    
    # add a horizontal line
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)



# iterateion for other
for index in range(2,20,3):
    # keep track of the list index
    cat_index = int((index-2)/3)
    # add the subplot
    ax = fig.add_subplot(6,3, index+1)
    # plot the individual data
    ax.plot(women_degrees["Year"], women_degrees[other_cats[cat_index]], c=cb_dark_blue, linewidth=3, label="Women")
    ax.plot(women_degrees["Year"], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    ax
    # set limits
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    # set title
    ax.set_title(other_cats[cat_index])
    # set text annotations
    if index == 2:
        ax.text(2002, 75, "Women")
        ax.text(2002, 23, "Men")
    if index == 17:
        ax.text(2002, 68, "Men")
        ax.text(2002, 28, "Women")
    # remove all spines
    # hiding spines using the axis.spines.Spine class
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    # removing the axis labels except the bottom-most one
    if index < 15:
        ax.tick_params(bottom=False, top=False, right=False, left=False, labelbottom=False)
    else:
        ax.tick_params(bottom=False, top=False, right=False, left=False)
        ax.spines["bottom"].set_visible(True)
    # disable the y-labels 
    ax.tick_params(labelleft=False)
    
    # add a horizontal line
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)

plt.savefig("gender_degrees.png")
plt.show()
```
